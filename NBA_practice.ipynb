```python
def initialize_data(Mstat, Tstat):
    new_Mstat = Mstat.drop(['Rk', 'Arena'], axis=1)
    new_Tstat = Tstat

    team_stats1 = pd.merge(new_Mstat, new_Tstat, how='left', on='Team')
    return team_stats1.set_index('Team', inplace=False, drop=True)
```

```python
initialize_data(Mstat, Overall)
```

```python
def get_elo(team):
    try:
        return team_elo[team]
    except:
        # assign ori elo value if there is no elo value at start
        team_elo[team] = ori_elo
        return team_elo[team]
```

```python
# calculate elo for each team
def calc_elo(win_team, lose_team):
    winner_rank = get_elo(win_team)
    loser_rank = get_elo(lose_team)

    rank_diff = winner_rank - loser_rank
    exp = (rank_diff  * -1) / 400
    odds = 1 / (1 + math.pow(10, exp))
    # modify K value based on rank
    if winner_rank < 2100:
        k = 32
    elif winner_rank >= 2100 and winner_rank < 2400:
        k = 24
    else:
        k = 16

    # update rank
    new_winner_rank = round(winner_rank + (k * (1 - odds)))      
    new_loser_rank = round(loser_rank + (k * (0 - odds)))
    return new_winner_rank, new_loser_rank
```

```python
def  build_dataSet(all_data):
    print("Building data set..")
    X = []
    skip = 0
    for index, row in all_data.iterrows():

        Wteam = row['WTeam']
        Lteam = row['LTeam']

        #get initial elo value
        team1_elo = get_elo(Wteam)
        team2_elo = get_elo(Lteam)

        # add 100 for host team
        if row['WLoc'] == 'H':
            team1_elo += 100
        else:
            team2_elo += 100

        # put elo as the first value of the team
        team1_features = [team1_elo]
        team2_features = [team2_elo]

        # æ·»åŠ æˆ‘ä»¬ä»Žbasketball reference.comèŽ·å¾—çš„æ¯ä¸ªé˜Ÿä¼çš„ç»Ÿè®¡ä¿¡æ¯
        for key, value in team_stats.loc[Wteam].iteritems():
            team1_features.append(value)
        for key, value in team_stats.loc[Lteam].iteritems():
            team2_features.append(value)

        # å°†ä¸¤æ”¯é˜Ÿä¼çš„ç‰¹å¾å€¼éšæœºçš„åˆ†é…åœ¨æ¯åœºæ¯”èµ›æ•°æ®çš„å·¦å³ä¸¤ä¾§
        # å¹¶å°†å¯¹åº”çš„0/1èµ‹ç»™yå€¼
        if random.random() > 0.5:
            X.append(team1_features + team2_features)
            y.append(0)
        else:
            X.append(team2_features + team1_features)
            y.append(1)

        if skip == 0:
            print('X',X)
            skip = 1

        # æ ¹æ®è¿™åœºæ¯”èµ›çš„æ•°æ®æ›´æ–°é˜Ÿä¼çš„eloå€¼
        new_winner_rank, new_loser_rank = calc_elo(Wteam, Lteam)
        team_elo[Wteam] = new_winner_rank
        team_elo[Lteam] = new_loser_rank

    return np.nan_to_num(X), y
```

```python
if __name__ == '__main__':

    #Mstat = pd.read_csv(folder + '/15-16Miscellaneous_Stat.csv')
    #Ostat = pd.read_csv(folder + '/15-16Opponent_Per_Game_Stat.csv')
    #Tstat = pd.read_csv(folder + '/15-16Team_Per_Game_Stat.csv')
    Mstat = pd.read_csv('data/15-16Miscellaneous_Stat.csv')
    #Ostat = pd.read_csv('data 2/15-16Opponent_Per_Game_Stat.csv')
    Tstat = Overall

    team_stats = initialize_data(Mstat, Tstat)

    result_data = pd.read_csv('data/2015-2016_result.csv')
    X, y = build_dataSet(result_data)

    # è®­ç»ƒç½‘ç»œæ¨¡åž‹
    print("Fitting on %d game samples.." % len(X))

    model = linear_model.LogisticRegression(penalty='l2',dual=False,tol=0.0001,C=1.0,fit_intercept=True,intercept_scaling=1,class_weight=None,random_state=None,solver='liblinear',max_iter=100,multi_class='ovr',verbose=0,warm_start=False,n_jobs=1)
#    model = linear_model.LogisticRegression(random_state=10, solver='liblinear')
    
    #penalty='l2',dual=False,tol=0.0001,C=1.0,fit_intercept=True,intercept_scaling=1,class_weight=None,random_state=None,solver='liblinear',max_iter=100,multi_class='ovr',verbose=0,warm_start=False,n_jobs=1

    model.fit(X, y)

    # åˆ©ç”¨10æŠ˜äº¤å‰éªŒè¯è®¡ç®—è®­ç»ƒæ­£ç¡®çŽ‡
    print("Doing cross-validation..")
    print(cross_val_score(model, X, y, cv = 5, scoring='accuracy', n_jobs=-1).mean())
```

```python
def predict_winner(team_1, team_2, model):
    features = []

    # team 1ï¼Œå®¢åœºé˜Ÿä¼
    features.append(get_elo(team_1))
    for key, value in team_stats.loc[team_1].iteritems():
        features.append(value)

    # team 2ï¼Œä¸»åœºé˜Ÿä¼
    features.append(get_elo(team_2) + 100)
    for key, value in team_stats.loc[team_2].iteritems():
        features.append(value)

    features = np.nan_to_num(features)
    return model.predict_proba([features])
```

```python
print('Predicting on new schedule..')
schedule1617 = pd.read_csv('data/16-17Schedule.csv')
result = []
for index, row in schedule1617.iterrows():
    team1 = row['Vteam']
    team2 = row['Hteam']
    pred = predict_winner(team1, team2, model)
    prob = pred[0][0]
    if prob > 0.5:
        winner = team1
        loser = team2
        result.append([winner, loser, prob])
    else:
        winner = team2
        loser = team1
        result.append([winner, loser, 1 - prob])

with open('16-17Result.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['win', 'lose', 'probability'])
    writer.writerows(result)
    print('done.')
```
